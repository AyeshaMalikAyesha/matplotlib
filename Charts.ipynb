# Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### **Gas Price Data Over Time in USD**

### Line Graph

```python
gas=pd.read_csv('gas_prices.csv')
plt.plot(gas.Year,gas.USA,'b^--',label='USA')
plt.plot(gas.Year,gas.Canada,'r.-',label='Canada')
plt.plot(gas.Year,gas['South Korea'],'g.-',label='South Korea')
plt.plot(gas.Year,gas.Australia,'y.-',label='Australia')

plt.title('Gas Prices over Time in USD',fontdict={'color':'brown'})

plt.xlabel('Year')
plt.ylabel('US Dollars($)')

plt.xticks(gas.Year[::3])

plt.legend()

plt.show()
```

```python
gas=pd.read_csv('gas_prices.csv')

for country in gas:
  if country!='Year':
    plt.plot(gas.Year,gas[country],label=country,marker='.')

plt.title('Gas Prices Over Time in USD',fontdict={'color':'silver'})
plt.xlabel('Year')
plt.ylabel('Price in USD')
plt.xticks(gas.Year[::3].tolist()+[2011])
plt.legend()
plt.savefig('gas_price_chart.png',dpi=300)
plt.show()
```

### **FIFA DATA**

```python
fifa=pd.read_csv('fifa_data.csv')
```

**Histogram**

```python
plt.hist(fifa.Overall,bins=50,color='#ab10bf')

plt.xlabel('Skill Level')
plt.ylabel('Number of Players')

plt.title('Distribution of player skills in FIFA 2018')

plt.show()
```

**Pie Charts**

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels=['Left','Right']
colors=['#abcdee','#abffbf']

plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f%%')

plt.title('Foot Preference of fifa palyers')

plt.show()
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')


light=fifa.loc[fifa.Weight<125].count()[0]
light_medium=fifa.loc[(fifa.Weight>=125) & (fifa.Weight<150)].count()[0]
medium=fifa.loc[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
medium_heavy=fifa.loc[(fifa.Weight>=175) & (fifa.Weight<200)].count()[0]
heavy=fifa.loc[fifa.Weight>=200].count()[0]

weights=[light,light_medium,medium,medium_heavy,heavy]
labels=['Under 125','125-150','150-175','175-200','Over 200']
plt.pie(weights,labels=labels,autopct='%.2f%%',pctdistance=0.9)


plt.show()
```

**Box plot**

```python
plt.style.use('default')

plt.figure(figsize=(7,10))

barcelona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
revs=fifa.loc[fifa.Club=='New England Revolution']['Overall']

boxes=plt.boxplot([barcelona,madrid,revs],labels=['Barcelona','Madrid','revs'],patch_artist=True)


for box in boxes['boxes']:
  box.set(color='#4289fe',linewidth=2)

  box.set(facecolor='#e0e0e0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
```
